```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import cycle

from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, label_binarize
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, auc
from sklearn.ensemble import VotingClassifier
# Ensemble models and base classifiers
from xgboost import XGBClassifier
from sklearn.ensemble import GradientBoostingClassifier, BaggingClassifier, RandomForestClassifier, ExtraTreesClassifier
from catboost import CatBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
import warnings
warnings.filterwarnings("ignore")

# ------------------------------
# 1. Load Dataset and Initial EDA
# ------------------------------
df = pd.read_csv("ObesityDataSet.csv")
print("Dataset Info:")
print(df.info())
print("\nFirst five rows:")
print(df.head())

# Plot histograms for numeric features (adjust column list if needed)
numeric_cols = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']
df[numeric_cols].hist(bins=20, figsize=(15,10))
plt.suptitle("Histograms for Numeric Attributes")
plt.show()

plt.figure(figsize=(12,10))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix for Numeric Features")
plt.show()

# ------------------------------
# 2. Data Preprocessing
# ------------------------------
# a) Missing Value Imputation: fill numeric with median, categorical with mode.
for col in df.columns:
    if df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        df[col].fillna(df[col].median(), inplace=True)

# b) Outlier Detection & Clipping using IQR on selected numeric columns
def clip_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[column] = np.clip(data[column], lower_bound, upper_bound)
    return data

# Example: clip outliers for 'Age' and 'NCP'
df = clip_outliers_iqr(df, 'Age')
df = clip_outliers_iqr(df, 'NCP')

# c) Label Encoding for categorical features.
# Based on your df.head() output, these columns are categorical:
categorical_cols = ['Gender', 'family_history_with_overweight', 'FAVC', 'CAEC', 'SMOKE', 'SCC', 'CALC', 'MTRANS']
le_dict = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    le_dict[col] = le

# d) Encode target variable ("NObeyesdad")
target_le = LabelEncoder()
df['NObeyesdad'] = target_le.fit_transform(df['NObeyesdad'])

# e) Feature Scaling: scale all predictor features using MinMaxScaler.
feature_cols = [col for col in df.columns if col != 'NObeyesdad']
scaler = MinMaxScaler()
df[feature_cols] = scaler.fit_transform(df[feature_cols])

# ------------------------------
# 3. Train-Test Split
# ------------------------------
X = df.drop('NObeyesdad', axis=1)
y = df['NObeyesdad']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# ------------------------------
# 4. Manual Implementation of Recursive Feature Elimination (RFE)
# ------------------------------
def manual_rfe(estimator, X, y, num_features_to_select):
    """
    Recursively eliminate the least-important feature based on the estimator's feature_importances_.
    Returns the ranking as a dictionary and the list of selected features.
    """
    remaining_features = list(X.columns)
    ranking = {}  # will store the order of removal
    iteration = 1
    while len(remaining_features) > num_features_to_select:
        estimator.fit(X[remaining_features], y)
        # Obtain feature importances (if the estimator doesn't provide it, adjust accordingly)
        importances = estimator.feature_importances_
        # Identify the feature with the lowest importance
        min_index = np.argmin(importances)
        feature_to_remove = remaining_features[min_index]
        ranking[feature_to_remove] = iteration
        print(f"Iteration {iteration}: Removing feature '{feature_to_remove}' (importance = {importances[min_index]:.4f})")
        remaining_features.remove(feature_to_remove)
        iteration += 1
    # Rank remaining features with the last iteration value
    for feat in remaining_features:
        ranking[feat] = iteration
        iteration += 1
    return ranking, remaining_features

# Use an XGBClassifier with parameters similar to the paper for RFE
base_estimator = XGBClassifier(learning_rate=0.1, n_estimators=100, max_depth=5,
                               min_child_weight=6, reg_alpha=60.0, subsample=0.6,
                               colsample_bytree=0.8, gamma=4.20, use_label_encoder=False,
                               eval_metric='mlogloss', random_state=42)

# Suppose we want to select 9 features (as the paper eventually used 9 after dropping negligible ones)
ranking, selected_features = manual_rfe(base_estimator, X_train, y_train, num_features_to_select=9)
print("\nFeature Ranking (lower number = removed earlier):")
print(ranking)
print("\nFinal selected features from manual RFE:")
# Final selected features as per base paper
# Final selected features as per base paper with corrected column names
selected_features = ['Weight', 'Height', 'Age', 'FCVC', 'FAVC', 'family_history_with_overweight', 'NCP', 'CAEC', 'CH2O']
print("\nFinal selected features (as per base paper):")
print(selected_features)


# Reduce training and test sets to the selected features
X_train_sel = X_train[selected_features]
X_test_sel = X_test[selected_features]

# ------------------------------
# 5. Hyperparameter Tuning with 10-fold Cross Validation
# ------------------------------
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# Define parameter grids (inspired by the paper's Table 5)
param_grids = {
    'XGB': {
    'learning_rate': [0.01, 0.05, 0.1],
    'n_estimators': [1000, 5000, 10000],
    'max_depth': [3, 5, 7, 10],
    'min_child_weight': [1, 6],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.7, 0.8, 0.9],
    'gamma': [0, 1, 4.2],
    'reg_alpha': [0, 0.1, 60.0],
    'reg_lambda': [0, 1, 10]
    },

    'GB': {
    'learning_rate': [0.01, 0.05, 0.1],
    'n_estimators': [5000, 10000],
    'max_depth': [3, 5, 7],
    'subsample': [0.7, 0.8, 1.0],
    'min_samples_split': [2, 10],
    'min_samples_leaf': [1, 2]
    },

    'CB': {
    'learning_rate': [0.05, 0.1, 0.2],
    'n_estimators': [1000, 5000],
    'max_depth': [2, 5, 10],
    'min_child_samples': [50, 250],
    'l2_leaf_reg': [1, 3, 10],
    'random_state': [45],
    'verbose': [1]
    },

    'RF': {
    'n_estimators': [500, 1000],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 10],
    'min_samples_leaf': [1, 2],
    'max_features': ['auto', 10, 16],
    'random_state': [42],
    'bootstrap': [True, False]
    },

    'ET': {
    'n_estimators': [500, 1000],
    'max_depth': [None, 10, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2],
    'max_features': ['auto', 10],
    'bootstrap': [True, False],
    'random_state': [42]
    },

    'BDT': {
    'n_estimators': [500, 1000],
    'base_estimator': [None, DecisionTreeClassifier(max_depth=3)],
    'bootstrap_features': [True, False],
    'oob_score': [False],
    'random_state': [42],
    'verbose': [0]
    }
}




# Instantiate base models (without tuning for Voting; use default for demonstration)
xgb_model = XGBClassifier(eval_metric='mlogloss', random_state=42)
gb_model = GradientBoostingClassifier(random_state=45)
cb_model = CatBoostClassifier(silent=True, random_state=45)
rf_model = RandomForestClassifier(bootstrap=True, random_state=42)
et_model = ExtraTreesClassifier(random_state=42)
bdt_model = BaggingClassifier(random_state=42, n_jobs=-1)

models_to_tune = {
    'XGB': xgb_model,
    'GB': gb_model,
    'CB': cb_model,
    'RF': rf_model,
    'ET': et_model,
    'BDT': bdt_model
}

best_estimators = {}

for name, model in models_to_tune.items():
    print(f"\nTuning model: {name}")
    grid = param_grids.get(name, {})
    if grid:
        gs = GridSearchCV(estimator=model, param_grid=grid, cv=cv, scoring='accuracy', n_jobs=-1)
        gs.fit(X_train_sel, y_train)
        best_estimators[name] = gs.best_estimator_
        print(f"Best parameters for {name}: {gs.best_params_}")
        print(f"Best CV accuracy for {name}: {gs.best_score_:.4f}")
    else:
        model.fit(X_train_sel, y_train)
        best_estimators[name] = model

# For the Voting ensemble, we combine Logistic Regression, Decision Tree, and SVC.
lr_model = LogisticRegression(random_state=42, max_iter=1000)
dt_model = DecisionTreeClassifier(random_state=42)
svc_model = SVC(C=2, probability=True, random_state=42)
voting_model = VotingClassifier(estimators=[('lr', lr_model), ('dt', dt_model), ('svc', svc_model)],
                                voting='soft')
voting_model.fit(X_train_sel, y_train)
best_estimators['Voting'] = voting_model

# ------------------------------
# 6. Evaluation on Test Data
# ------------------------------
# For multiclass ROC-AUC, binarize the target labels.
y_test_binarized = label_binarize(y_test, classes=sorted(y.unique()))
n_classes = y_test_binarized.shape[1]

def plot_multiclass_roc(model, X_test, y_test, n_classes, title):
    y_score = model.predict_proba(X_test)
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_binarized[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
    # Compute micro-average ROC curve and ROC area
    fpr["micro"], tpr["micro"], _ = roc_curve(y_test_binarized.ravel(), y_score.ravel())
    roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])
   
    plt.figure(figsize=(8, 6))
    plt.plot(fpr["micro"], tpr["micro"],
             label='micro-average ROC curve (area = {0:0.2f})'.format(roc_auc["micro"]),
             color='deeppink', linestyle=':', linewidth=4)
   
    colors = cycle(['aqua', 'darkorange', 'cornflowerblue', 'red', 'green', 'purple', 'brown'])
    for i, color in zip(range(n_classes), colors):
        plt.plot(fpr[i], tpr[i], color=color, lw=2,
                 label='ROC curve of class {0} (area = {1:0.2f})'.format(i, roc_auc[i]))
   
    plt.plot([0, 1], [0, 1], 'k--', lw=2)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(title)
    plt.legend(loc="lower right")
    plt.show()

# Evaluate each tuned model
for name, model in best_estimators.items():
    print(f"\n--- Evaluation for {name} Model ---")
    y_pred = model.predict(X_test_sel)
   
    # Print classification report (decode target labels for readability)
    target_names = target_le.inverse_transform(sorted(y.unique()))
    print(f"Classification Report for {name}:")
    print(classification_report(y_test, y_pred, target_names=target_names))
   
    # Plot confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"Confusion Matrix - {name}")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()
   
    # ROC-AUC evaluation (if model supports predict_proba)
    if hasattr(model, "predict_proba"):
        y_pred_proba = model.predict_proba(X_test_sel)
        roc_auc = roc_auc_score(y_test_binarized, y_pred_proba, average='macro', multi_class='ovr')
        print(f"Macro-average ROC-AUC for {name}: {roc_auc:.2f}")
        plot_multiclass_roc(model, X_test_sel, y_test, n_classes, f"ROC Curves - {name}")
    else:
        print(f"{name} does not support probability estimates for ROC evaluation.")

import joblib

# Save each trained model to disk
for name, model in best_estimators.items():
    filename = f"best_{name}_model.pkl"
    joblib.dump(model, filename)
    print(f"Saved {name} model to {filename}")

# Example: To load a model later:
# loaded_model = joblib.load("best_XGB_model.pkl")
```
